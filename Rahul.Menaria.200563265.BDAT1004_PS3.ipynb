## Question 1

```python
import pandas as pd
import numpy as np

data_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
user_data = pd.read_csv(data_url, sep='|')

average_age_by_job = user_data.groupby('occupation')['age'].mean()
male_proportion_by_job = user_data[user_data['gender'] == 'M'].groupby('occupation')['gender'].count() / user_data.groupby('occupation')['gender'].count()
sorted_male_proportion = male_proportion_by_job.sort_values(ascending=False)
age_range_by_job = user_data.groupby('occupation')['age'].agg(['min', 'max'])
average_age_by_job_and_gender = user_data.groupby(['occupation', 'gender'])['age'].mean()
gender_distribution_by_job = user_data.groupby(['occupation', 'gender'])['gender'].count() / user_data.groupby('occupation')['gender'].count() * 100

print("Average Age per Occupation:")
print(average_age_by_job)

print("\nMale Proportion per Occupation (Sorted):\n")
print(sorted_male_proportion)

print("\nAge Range per Occupation:\n")
print(age_range_by_job)

print("\nAverage Age per Occupation and Gender:\n")
print(average_age_by_job_and_gender)

print("\nGender Distribution per Occupation:\n")
print(gender_distribution_by_job)
```

## Question 2

```python
import pandas as pd
import numpy as np

data_url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
euro_data = pd.read_csv(data_url)

total_goals = euro_data['Goals']
num_teams = euro_data.shape[0]  # Number of rows in the dataframe
num_columns = euro_data.shape[1]  # Number of columns in the dataframe

discipline_data = euro_data[['Team', 'Yellow Cards', 'Red Cards']]
sorted_discipline = discipline_data.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=[False, False])

avg_yellow_cards = discipline_data.groupby('Team')['Yellow Cards'].mean()
high_scoring_teams = euro_data[euro_data['Goals'] > 6]
teams_starting_with_G = euro_data[euro_data['Team'].str.startswith('G')]

first_seven_columns = euro_data.iloc[:, :7]
all_but_last_three_columns = euro_data.iloc[:, :-3]

shooting_accuracy = euro_data[euro_data['Team'].isin(['England', 'Italy', 'Russia'])][['Team', 'Shooting Accuracy']]

print("Number of teams participated in Euro 2012:", num_teams)
print("Number of columns in the dataset:", num_columns)

print("\nTop 5 rows of the Euro 2012 dataset:")
print(euro_data.head())

print("\nGoals column:")
print(total_goals)

print("\nDiscipline dataframe:")
print(discipline_data)

print("\nTeams sorted by Red Cards and Yellow Cards:")
print(sorted_discipline)

print("\nAverage Yellow Cards per Team:")
print(avg_yellow_cards)

print("\nTeams that scored more than 6 goals:")
print(high_scoring_teams)

print("\nTeams starting with 'G':")
print(teams_starting_with_G)

print("\nFirst 7 columns:")
print(first_seven_columns)

print("\nAll columns except the last 3:")
print(all_but_last_three_columns)

print("\nShooting Accuracy for England, Italy, and Russia:")
print(shooting_accuracy)
```

```python
#Assign it to a variable called data and replace the first 3 columns by a proper datetime index
data["Date"] = pd.to_datetime(data[["Yr","Mo","Dy"]].astype(str).agg('-'.join,axis=1))
data = data.drop(columns = ["Yr","Mo","Dy"])
data = data[['Date', 'RPT', 'VAL', 'ROS', 'KIL', 'SHA', 'BIR', 'DUB', 'CLA', 'MUL', 'CLO', 'BEL', 'MAL']]
data.head()
```

```python
#Year 2061? Do we really have data from this year? Create a function to fix it and apply it
NData = data.set_index("Date")
#Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
NData.index.astype("datetime64[ns]")
```

```python

# . Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below
NData.isnull().values.ravel().sum()
```

```python
#. Compute how many non-missing values there are in total.
nData = NData.count()
print("Total non-missing values are :", nData.sum())
```

```python
#Calculate the mean windspeeds of the windspeeds over all the locations and all the times. A single number for the entire dataset.
x = round(NData.mean(), 2)
x
```

```python
#Single number for the entire dataset
round(x.mean(),2)
```

```python
#Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days A different set of numbers for each location.
#Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day. A different set of numbers for each day.
def statsKPI(x):
    x = pd.Series(x)
    Mean = x.mean()
    Std = x.std()
    Min = x.min()
    Max = x.max()
    res = [Min,Max,Mean,Std]
    indx = ["MEAN", "Std", "Min", "Max"]
    result = pd.Series(res,index = indx)
    return result
```

```python
#A different set of numbers for each location
loc_stats = NData.apply(statsKPI)
loc_stats
```

```python
#Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at eachday.A different set of numbers for each day
day_stats = NData.apply(statsKPI, axis = 1)
day_stats.head()
```

```python

#Find the average windspeed in January for each location. Treat January 1961 and January 1962 both as January
jan_AVG_wspeed = NData[NData.index.month ==1]
round(jan_AVG_wspeed.mean(),2)
```

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

# Step 4: See the first 10 entries
print(chipo.head(10))

# Step 5: Number of observations
num_observations = len(chipo)

# Step 6: Number of columns
num_columns = len(chipo.columns)

# Step 7: Print column names
column_names = chipo.columns

# Step 8: Index type
index_type = chipo.index

# Step 9: Most-ordered item
most_ordered_item = chipo['item_name'].value_counts().idxmax()

# Step 10: Number of most-ordered item
most_ordered_item_count = chipo['item_name'].value_counts().max()

# Step 11: Most ordered item in choice_description column
most_ordered_choice = chipo['choice_description'].value_counts().idxmax()

# Step 12: Total items ordered
total_items_ordered = chipo['quantity'].sum()

# Step 13: Convert item price to float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

# Step 14: Revenue for the period
revenue = chipo['item_price'].sum()

# Step 15: Number of orders
num_orders = chipo['order_id'].nunique()

# Step 16: Average revenue per order
avg_revenue_per_order = revenue / num_orders

# Step 17: Number of different items sold
num_different_items = chipo['item_name'].nunique()

# Print the formatted results
print("\nNumber of observations:", num_observations)
print("Number of columns:", num_columns)
print("Column names:", column_names)
print("Index type:", index_type)
print("Most-ordered item:", most_ordered_item)
print("Number of most-ordered item:", most_ordered_item_count)
print("Most ordered choice description:", most_ordered_choice)
print("Total items ordered:", total_items_ordered)
print("Revenue for the period: ${:.2f}".format(revenue))
print("Number of orders:", num_orders)
print("Average revenue per order: ${:.2f}".format(avg_revenue_per_order))
print("Number of different items sold:", num_different_items)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the file
file_path = "us-marriages-divorces-1867-2014.csv"  # Replace with the actual file path
marriage_data = pd.read_csv(file_path)

# Filter data for the years 1900, 1950, and 2000
selected_years = [1900, 1950, 2000]
data_subset = marriage_data[marriage_data['Year'].isin(selected_years)]

# Create a vertical bar chart
plt.figure(figsize=(10, 6))

bar_width = 5
plt.bar(data_subset['Year'], data_subset['Marriages_per_1000'], width=bar_width, label='Marriages per 1000', color='lightblue')
plt.bar(data_subset['Year'] + bar_width, data_subset['Divorces_per_1000'], width=bar_width, label='Divorces per 1000', color='orange')

plt.title('U.S. Marriages and Divorces per Capita (1900, 1950, 2000)')
plt.xlabel('Year')
plt.ylabel('Rate per 1000')
plt.xticks(data_subset['Year'] + bar_width / 2, selected_years)
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("roman-emperor-reigns.csv")

# Count the number of emperors based on their cause of death
num_assassinated_emperors = data[data['Cause_of_Death'] == 'Assassinated']['Emperor'].count()
num_not_assassinated_emperors = len(data) - num_assassinated_emperors

# Prepare data for the pie chart
death_categories = ['Assassinated', 'Not Assassinated']
emperor_counts = [num_assassinated_emperors, num_not_assassinated_emperors]
colors = ['#ff6347', '#3cb371']  # Updated colors
explode = (0.1, 0)  # Highlight the 'Assassinated' slice

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(emperor_counts, explode=explode, labels=death_categories, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Fraction of Roman Emperors Assassinated')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular

plt.show()
```
